<h2> TASK 4   : Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. </h2>

<h2>Goal_diff_count = GoalsFor - GoalsAgainst </h2>

```python
def Goal_diff_count():
    my_data=data[['Team','GoalsFor','GoalsAgainst']]
    my_data=my_data.reset_index(drop=True)
    my_data['Goal_diff_count'] = data['GoalsFor'] - data['GoalsAgainst']
    maxi=my_data['Goal_diff_count'].max()
    mini=my_data['Goal_diff_count'].min()
    print(my_data)
    print(f'Maximum Goal Difference is {maxi}')
    print(f'Minimum Goal Difference is {mini}')
```

```python
Goal_diff_count()
```
